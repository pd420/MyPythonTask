### What is a Data Structure? 

Organizing, managing and storing data is important as it enables easier access and efficient modifications. Data Structures allows you to organize your data in such a way that enables you to store collections of data, relate them and perform operations on them accordingly. 

### Types of Data Structures in Python

Python has implicit support for Data Structures which enable you to store and access data. These structures are called List, Dictionary, Tuple and Set.

Python allows its users to create their own Data Structures enabling them to have full control over their functionality. The most prominent Data Structures are Stack, Queue, Tree, Linked List and so on which are also available to us in other programming languages.

### Built-in Data Structures

Python has four basic inbuilt data structures namely Lists, Dictionary, Tuple and Set.

#### 1)List

Lists are used to store data of different data types in a sequential manner. There are addresses assigned to every element of the list, which is called as Index. The index value starts from 0 and goes on until the last element called the positive index. There is also negative indexing which starts from -1 enabling you to access elements from the last to first. 

Creating a list

To create a list, we use the square brackets and add elements into it accordingly. If we do not pass any elements inside the square brackets, we get an empty list as the output.

```python
my_list = [] #create empty list
my_list
```

```python
my_list = [1, 2, 3, 'example', 3.132] #creating list with data
my_list
```

There are various Functions that can be carried out on lists like append(), extend(), reverse(), pop() etc. 

#### 2)Dictionary

In python, dictionary is similar to hash or maps in other languages. It consists of key value pairs. The value can be accessed by unique key in the dictionary. Keys are unique & immutable objects.

Creating a Dictionary

Dictionaries can be created using the flower braces or using the dict() function. We need to add the key-value pairs whenever we work with dictionaries.

Syntax:

dictionary = {"key name": value}


```python
my_dict = {} #empty dictionary
my_dict
```

```python
my_dict = {1: 'Python', 2: 'Java'} #dictionary with elements
my_dict
```

#### 3)Tuple

Tuples are the same as lists are with the exception that the data once entered into the tuple cannot be changed no matter what. The only exception is when the data inside the tuple is mutable, only then the tuple data can be changed.

Creating a Tuple

We will create a tuple using parenthesis or using the tuple() function.



```python
my_tuple = (1, 2, 3) #create tuple
my_tuple 
```

```python
#() An empty tuple
t1 = (0, ) #A one-item tuple
t1
```

#### 4)Sets

Sets are a collection of unordered elements that are unique. Meaning that even if the data is repeated more than one time, it would be entered into the set only once.

Creating a set

Sets are created using the flower braces but instead of adding key-value pairs, we just pass values to it.

```python
my_set = {1, 2, 3, 4, 5, 5, 5} #create set
my_set
```
